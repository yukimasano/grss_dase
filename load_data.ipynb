```python
import scipy.io as sio
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline

import torch
import torchvision
import torchvision.transforms as transforms
```

```python
pathToData = '/scratch/vary/data/rit18/rit18_data.mat'
mat_contents = sio.loadmat(pathToData)
```

```python
train_data = mat_contents['train_data'].astype('int16',casting ='same_kind')
train_labels = mat_contents['train_labels'].astype('int16',casting ='same_kind')
```

```python
train_M = train_data.shape[1]
train_N = train_data.shape[2]
train_B = train_data.shape[0]

train_data = train_data.reshape(7, train_M*train_N)
train_data = train_data.transpose()
train_labels = train_labels.reshape(-1)
```

```python
plt.matshow(train_data[:,4].reshape(train_M, train_N))
plt.show()
plt.imshow(train_labels.reshape(train_M, train_N))
plt.show()
```
